A lot of information received is generally stored in form of tables and text and we can access the exact amount or numbers that respond or correspond to the information but then why do we need to visualize it.
The answer to this question lies in the inner workings of the human mind and it is a well-known fact that our brain recognizes images faster. At birth, a baby understands the image of his or her mother months before being able to understand what the word mommy means. We understand images instantly and we have to work to process text.
In fact, the brain processes image 60,000 times faster than it does text. And it’s more accustomed to processing images—ninety percent of the information sent to the brain is visual, and 93 % of all human communication is visual and none of this is new or recent. The human brain has always processed images ridiculously faster than words.
Then it certainly means a lot to visualize data and build stories around about it to explain them intuitively and thus enables us to derive insight out of visual information and that is why we need visualization as important part of the process while dealing with data providing insights that can help us take effective business decisions.

```python
#Importing Python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading Data
data1 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='Sheet1')
data2 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='Sheet2')
data3 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='Sheet3')
data4 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='sp_data')
data5 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='boxp_data')
data6 = pd.read_excel("UE_Trend Investigation.xlsx", sheet_name='line_data')
```

### LINE GRAPH
A line chart or line graph is a type of chart which displays information as a series of data points called ‘markers’ connected by straight line segments. It is a basic type of  chart common in many fields. It is similar to a scatter plot except that the measurement points are ordered (typically by their x-axis value) and joined with straight line segments. A line chart is often used to visualize a trend in data over intervals of time

```python
data6.head(3)
```

```python
#Line BarGraph

ax=data6["Store Count_2016"].plot(marker='o', markerfacecolor='red', markersize=12, color='black', linewidth=4,legend=True)
ax=data6["Store Count_2017"].plot(marker='o', markerfacecolor='blue', markersize=12, color='yellow', linewidth=4,legend=True)
ax.set_xticklabels(data6["REGION"], rotation=90)

plt.xlabel('Region', fontsize=15)
plt.ylabel('Store Count', fontsize=15)
plt.title('Title')
#plt.xticks(data6.index)
plt.show()
#Insight-Trend of store Counts in 2 waves
```

```python
Matplotlib automatically coloured the lines for us, but we can control this. 


Starting with colours, the following abbreviations are supported:

*Blue – ‘b’
*Green – ‘g’
*Red – ‘r’
*Cyan – ‘c’
*Magenta – ‘m’
*Yellow – ‘y’
*Black – ‘k’ (‘b’ is taken by blue so the last letter is used)
*White – ‘w’

We specify line type as follows:

*Solid Line – ‘-‘
*Dashed – ‘–‘
*Dotted – ‘.’

...and we can set markers, such as:

*Point – ‘.’
*Pixel – ‘,’
*Circle – ‘o’
*Square – ‘s’
*Triangle – ‘^’
```

## SCATTER PLOT
A ScatterPlot displays the value of 2 sets of data on 2 dimensions. Each dot represents an observation. The position on the X (horizontal) and Y (vertical) axis represents the values of the 2 variables. It is really useful to study the relationship between both variables. It is common to provide even more information using colors or shapes

```python
data4.head()
```

```python
#Scatter Plot
plt.scatter(data4["Numeric distribution (%)"],data4["Actual SE (Used)"],marker="*",color="green")
plt.show()
```

```python
#Scatter Plot with title, axis label and legend

plt.scatter(data4["Numeric distribution (%)"],data4["Actual SE (Used)"],marker="*",color="Red")
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('ScatterPlot')
plt.savefig('sample.png')
plt.show()
#INSIGHT-These plots are used to understand whether our sample design is meeting precision or not.
#If there is any point in the plot which has numeric distribution higher than 80% and precision/SE higher than what is allowed for that MBD then we should investigate the case and check if the design meets the limit.
```

## HISTOGRAM
An histogram is an accurate graphical representation of the distribution of numerical data. It takes as input one numerical variable only. The variable is cut into several bins, and the number of observation per bin is represented by the height of the bar. Note that the shape of the histogram can be really different following the number of bins you set.

```python
data1.head(3)
```

```python
#Histogram
#plotting a histogram
plt.hist(data1["2017 W2"])
#plt.hist(data1["2017 W2"],range=[10,1000], bins=10, rwidth = 0.7)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Distribution of Store Count in 2017 W2')
plt.show()
#INSIGHT-80% of Regions have Store Count between 0 to 400
```

## BOXPLOT
 Box plot is probably one of the most common type of graphic. It gives a nice summary of one or several numeric variables. The line that divides the box into 2 parts represents the median of the data. The end of the box shows the upper and lower quartiles. The extreme lines shows the highest and lowest value excluding outliers. 

```python
data5.head(3)
```

```python
#Box Plot
plt.boxplot(data5["STORE_COUNT_2017"])
#plt.boxplot(data5["STORE_COUNT_2017"],patch_artist=False,labels=['STORE_COUNT_2017'])
plt.show()
#Insight-Median store count and outliers for 2017
```

```python
#Box Plot
plt.boxplot([data5["STORE_COUNT_2017"],data5["STORE_COUNT_2018"]],patch_artist=False,
            labels=['STORE_COUNT_2017','STORE_COUNT_2018'])
plt.show()
```

## PIE CHART
Pie chart is probably one of the most common type of chart. It is a circular graphic which is divided into slices to illustrate numerical proportion. The point of a pie chart is to show the relationship of parts out of a whole

```python
data3.head()
```

```python
#Pie_chart

# portion covered by each label
slices = data3["2016 W1 - Sign-off"]
label=["a","b","c","d","e","g","g","t"]

# color for each label
#colors = ['r', 'y', 'g', 'b']

# plotting the pie chart
plt.pie(slices,startangle=90,radius = 1.1 ,labels=label,explode=(0,0.1,0,0,0.5,0,0,0),autopct='%1.0f%%')
plt.legend(data3["Category"], bbox_to_anchor=(0, 1))

# Saving the Chart in local machine
plt.savefig('sample.pdf')
plt.show()
#INSIGHT-Share of each category in the total universe in 2016 W1 
```

## BARPLOT
A barplot is one of the most common type of plot. It shows the relationship between a numerical variable and a categorical variable. For example, you can display the height of several individuals using bar chart.

```python
data3.head()
```

```python
#BARGRAPH
ax=data3["2017 W1"].plot.bar()
ax.set_xticklabels(data3["Category"], rotation=45)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Title')
plt.show()
#INSIGHT:Total number of stores in each category in 2017 W1
```

## GROUP BARGRAPH

```python
ax=data3[["2017 W1","2017 W2"]].plot.bar()
ax.set_xticklabels(data3["Category"], rotation=45)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Title')
plt.show()
#Comparison of store count in each category in both waves
```

## STACKED BARGRAPHS
This section display grouped barcharts, stacked barcharts . This 2 types of barplot variation have the same objective. It displays a numerical value for several entities, organised into groups and subgroups. A grouped barplot display the subgroups one beside each other, whereas the stacked ones display them on top of each other.

```python
#Stacked BarGraph
ax=data3[["2017 W1","2017 W2"]].plot.bar(stacked=True)
ax.set_xticklabels(data3["Category"], rotation=45)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Title')
plt.show()
#Comparison of store count in each category in both waves
```

<div id="1.1" style="height:40px;width:1000px;opacity:0.8;display:inline-block;background-color:rgba(30,144,255,0.5);"><h2> SEABORN</h2></div> 

Seaborn is a graphic library built on top of Matplotlib. It allows to make your charts prettier, and facilitates some of the common data visualisation needs (like mapping a color to a variable or using faceting).For more details visit:https://python-graph-gallery.com/seaborn/

```python
import seaborn as sns                   # <--- import seaborn

sns.set()                               # <--- call this before a plot, to reset to defaults
```

```python
#Scatterplot
# use the function regplot to make a scatterplot
sns.regplot(data1["2016 W1"],data1["Trend"],fit_reg=False)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Trend in 2016')
plt.show()
```

```python
#HISTOGRAM
sns.distplot(data1["2016 W1"],bins=10)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.title('Histogram chart')
plt.show()
```

```python
#BARPLOT
ax=sns.barplot(x="Category", y="2017 W2", data=data3)
ax.set_xticklabels(data3["Category"], rotation=45)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)

plt.title('Barplot chart')
plt.show()
```

```python
#BOXPLOT
sns.boxplot( x=data5["STORE_COUNT_2017"],notch=True,width=0.3,orient="v",labels=['STORE_COUNT_2017'])
plt.show()
```
